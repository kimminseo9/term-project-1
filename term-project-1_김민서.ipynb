## 1-1. 주제
### 사이버 범죄의 유형과 현 상황

## 1-2. 주제 선정 이유

대한민국은 IT관련 기술을 발전시키며 어느새 IT강국으로 자리잡았다. 이러한 기술의 발전은 사람들에게 일상생활 속 편리성을 안겨주었다. 그러나, 이는 이와 동시에 위험성 또한 수반하였는데, 대표적인 것이 바로 '사이버 범죄'이다.
해당 프로젝트에서는 사이버 범죄의 유형과 발생건수를 알아보고, 해당 범죄의 발생건수에 대한 검거건수의 비율을 분석하고자 하며 이를 통해 앞으로 사이버 범죄를 예방하기 위해 필요한 방향성을 고찰하고자 한다.

## 2. 가설 정의


- 가설1. 사이버 범죄는 과거부터 현재까지 점점 증가했을 것이다. :시간이 지날수록 IT 기술이 발전하기 때문에 관련 범죄도 늘어날 것이라고 생각하였다.
- 가설2. 사이버 범죄의 발생건수는 월별로 큰 차이가 없을 것이다. :사이버 범죄는 날짜, 날씨 등에 큰 영향을 받지 않으므로 월별로 큰 차이가 없을 것이라고 생각하였다.
- 가설3. 사이버 범죄의 검거율(%)은 점점 증가했을 것이다. :시간이 지날수록 IT 기술이 발전하므로 이에 대한 검거율 또한 높을 것으로 예상하였다.
- 가설4. 사이버 범죄의 검거율(%)은 해킹이 가장 낮을 것이다. :해킹 관련 범죄는 뉴스나 기사로 자주 접했기 때문에 많이 발생하였을 것으로 생각하였고, 범죄의 양이 너무 많아 모든 사건에 대한 검거가 어려울 것으로 예상하였다.
- 가설5. 사이버 범죄의 검거율(%)은 사이버 금융 범죄가 가장 높을 것이다. :금융은 다른 분야보다 보안이 더욱 중요하기 때문에 검거율이 가장 높을 것이라고 생각하였다.

#### 가설을 확인하기 위한 최종 dataframe 만들기

```python
# 가설1: 사이버 범죄는 과거부터 현재까지 점점 증가했을 것이다.

df_가설1 = df_all1.drop([1, 3, 5, 7, 9, 11, 13]) # 기존 detaframe에서 검거건수를 제거하였다.
df_가설1 = df_가설1.reset_index(drop=True)

df_가설1
```

```python
# 가설 3: 사이버 범죄의 검거율(%)은 점점 증가했을 것이다.

df_가설3 = df_all1.copy()
for m in range(1,14,2):
    df_가설3.iloc[m-1,2] = (((df_all1.iloc[m,2])/(df_all1.iloc[m-1,2]))*100)
    df_가설3.iloc[m-1,1] = "검거율" # '발생건수'를 '검거율'로 바꾸었다.
df_가설3 = df_가설3.drop([1,3,5,7,9,11,13]) # 검거건수를 제거하였다.
df_가설3 = df_가설3.reset_index(drop=True) # rows의 인덱싱을 초기화하였다.
df_가설3 = df_가설3.rename(columns={'사이버범죄':'사이버범죄(%)'}) # '사이버 범죄'칼럼명을 '사이버 범죄(%)'로 바꾸어 단위를 표시하였다.

df_가설3
```

```python
# 가설4~5: 사이버 범죄의 검거율은 해킹이 가장 낮고, 사이버금융범죄가 가장 높을 것이다.

df_가설4_5 = df_all.copy()
for m in range(1,14,2):
    for s in range(2,11):
        divisor = df_all.iloc[m,s]
        if divisor != 0:
            df_가설4_5.iloc[m-1,s] = ((df_all.iloc[m,s])/(df_가설4_5.iloc[m-1,s]))*100
        else:
            df_가설4_5.iloc[m-1,s] = 100 # 분모(발생건수)가 0인 경우는 검거율 100%로 설정하였다.
    df_가설4_5.iloc[m-1,1] = "검거율"
df_가설4_5 = df_가설4_5.drop([1,3,5,7,9,11,13])
df_가설4_5 = df_가설4_5.reset_index(drop=True)
df_가설4_5 = df_가설4_5.rename(columns={'해킹':'해킹(%)', '서비스거부공격':'서비스거부공격(%)', '악성프로그램':'악성프로그램(%)', '정보범죄':'정보범죄(%)', '사이버사기':'사이버사기(%)', '사이버금융범죄':'사이버금융범죄(%)', '사이버음란물':'사이버음란물(%)', '사이버도박':'사이버도박(%)', '사이버명예훼손':'사이버명예훼손(%)'}) # 각각의 칼럼명에 단위를 표시하였다.

n=0
for u in range(2,11):
    for h in range(0,7):
        n += df_가설4_5.iloc[h,u]
    df_가설4_5.iloc[0,u] = (n/7) # 각 유형별로 사이버 범죄 검거율의 평균 %를 계산하였다.
    n=0
df_가설4_5 = df_가설4_5.drop([1,2,3,4,5,6]) # 첫번째 열을 제외한 열들을 모두 제거하였다.
df_가설4_5.iloc[0,0] = "평균"
df_가설4_5 = df_가설4_5.rename(columns={'연도':'방법'}) # '연도' 칼럼명을 '방법'으로 변경하였다.
    
df_가설4_5
```

### 월별 사이버 범죄 현황을 계산하기 위한 데이터 가공

```python
# 가설2: 사이버 범죄의 발생건수는 월별로 큰 차이가 없을 것이다.

df_가설2 = df_crime2.drop([1,3,5,7,9,11,13])
df_가설2 = df_가설2.reset_index(drop=True)
n=0
for u in range(2,14):
    for h in range(0,9):
        n += df_가설2.iloc[h,u]
    df_가설2.iloc[0,u] = (n/9).astype(int) # 각 유형별로 사이버 범죄건수의 평균을 계산하였다.
    n=0
df_가설2 = df_가설2.drop([1,2,3,4,5,6,7,8])
df_가설2.iloc[0,0] = "평균"
df_가설2 = df_가설2.rename(columns={'연도':'방법'})
df_가설2
```
